```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN, MeanShift, OPTICS
from sklearn.preprocessing import StandardScaler

# Veri setini yükle (Kaggle veya CSV)
df = pd.read_csv("Mall_Customers.csv")

# Özellik seçimi
X = df[['Annual Income (k$)', 'Spending Score (1-100)']]

# Ölçekleme
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# --------------------
# DBSCAN
# --------------------
dbscan = DBSCAN(eps=0.3, min_samples=5)
labels_dbscan = dbscan.fit_predict(X_scaled)

plt.figure()
plt.scatter(X_scaled[:,0], X_scaled[:,1], c=labels_dbscan)
plt.title("DBSCAN - Müşteri Segmentasyonu")
plt.xlabel("Gelir")
plt.ylabel("Harcama Puanı")
plt.show()

# --------------------
# Mean-Shift
# --------------------
meanshift = MeanShift()
labels_meanshift = meanshift.fit_predict(X_scaled)

plt.figure()
plt.scatter(X_scaled[:,0], X_scaled[:,1], c=labels_meanshift)
plt.title("Mean-Shift - Müşteri Segmentasyonu")
plt.xlabel("Income")
plt.ylabel("Spending Score")
plt.show()

# --------------------
# OPTICS
# --------------------
optics = OPTICS(min_samples=5)
labels_optics = optics.fit_predict(X_scaled)

plt.figure()
plt.scatter(X_scaled[:,0], X_scaled[:,1], c=labels_optics)
plt.title("OPTICS - Müşteri Segmentasyonu")
plt.xlabel("Income")
plt.ylabel("Spending Score")
plt.show()
```
